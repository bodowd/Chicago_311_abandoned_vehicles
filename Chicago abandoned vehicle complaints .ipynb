```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
```

```python
df = pd.read_csv('311_Service_Requests_-_Abandoned_Vehicles.csv')
```

```python
# Looking at what makes of vehicles have most reports of abandonment
makes_of_vehicles_counts = df['Vehicle Make/Model'].value_counts()
makes_of_vehicles = df['Vehicle Make/Model']

# Top 50 makes of abandoned vehicles
makes_of_vehicles_counts[0:50]
```

```python
# plot top 50 makes with 
fig, ax = plt.subplots(figsize=(10,10))
plt.tight_layout()



ax.tick_params(axis = 'y', labelsize = 12)
makes_of_vehicles_counts[0:50].plot.barh()
plt.gca().invert_yaxis()
```
